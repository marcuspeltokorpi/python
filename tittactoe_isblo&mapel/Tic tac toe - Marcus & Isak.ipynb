```python
from pandas import *
import pickle
import copy
import numpy
from numpy import zeros
```

```python
#Funktion för att skapa en lista med ett tomt bräde och en lista med ett "random" bräde.
def getEmptyBoard():
    return [0,0,0,0,0,0,0,0,0]

def getRandomBoard():
    return list(numpy.random.choice([0,1,2],(9)))

print(getEmptyBoard())
```

```python
#Metod för att skriva ut ett bräde med ".", "O" och "X" istället för 0, 1 och 2.
def showBoard(board):
    C=[".","O","X"]
    i=0
    for column in range(3):
        S=""
        for row in range(3):
            S=S+" "+C[board[i]]+" |"
            i=i+1
        print(S)
        print("---+---+---")
    print (" ")
    print (" ")
    
showBoard(getEmptyBoard())
showBoard(getRandomBoard())
print(" 0 1 2")
print(" 3 4 5")
print(" 6 7 8")
```

```python
#Denna koden definerar funktionen score(board) och tar brädan som input och retunerar vem som vunnit, 0 står för tie,
# 1 för player O och 2 för player X
def score(board):
#funktionen kontrollerar varje rad och kolumn och diagonalen av brädan för att kolla om den innehåller tre symboler 
#irad, alltså "O" eller "X", ifall det finns så sätts value till den spelaren som har det alltså 1 eller 2
    line=0
     #kollar rader
    if board[0]==board[1] and board[1]==board[2] and board[0]!=0:
        line=board[0]
    if board[3]==board[4] and board[4]==board[5] and board[3]!=0:
        line=board[3]
    if board[6]==board[7] and board[7]==board[8] and board[6]!=0:
        line=board[6]
    # kollar kolumner
    if board[0]==board[3] and board[3]==board[6] and board[0]!=0:
        line=board[0]
    if board[1]==board[4] and board[4]==board[7] and board[1]!=0:
        line=board[1]
    if board[2]==board[5] and board[5]==board[8] and board[2]!=0:
        line=board[2]
    # kollar diagonalen
    if board[0]==board[4] and board[4]==board[8] and board[0]!=0:
        line=board[0]
    if board[2]==board[4] and board[4]==board[6] and board[2]!=0:
        line=board[2]  
    return line #return 0 för tie eller inte klar, 1 = player O vinner, 2 = player X vinner

board=[1,1,1,0,0,0,2,2,2]
showBoard(board)
print(score(board))
```

```python
# generellt så är det denna kod som ser till att spelare kan göra ett drag och att det uppdateras på brädan.
def doMove(board,move,who):
    #deepcopy funktionen skapar en kopia av orginalbrädan och ser till så att den inte blir modifierad. 
    #Sen så sätter den value vid angivet drag till spelaren som gjorde sitt drag in i den nya kopian
    newBoard=copy.deepcopy(board)
    newBoard[move]=who
    return newBoard

board=getEmptyBoard()
showBoard(board)
board=doMove(board,4,1)
showBoard(board)
```

```python
#funktionen getAllMoves hittar alla möjliga drag genom att loopa genom brädan.
def getAllMoves(board):
    ret=[]
    for i in range(9):
        if board[i]==0:
            ret.append(i)
    return ret
# hämtar radombräde
board=getRandomBoard()
#skriver ut brädet
showBoard(board)
# printar alla möjliga drag på brädet.
print(getAllMoves(board))
```

```python
#En metod som använder sig av scoremetoden för att kolla om matchen är över
def gameOver(board):
    s=score(board)
    #Om antalet lediga rutor på brädet är noll eller om någon har vunnit så returneras True.
    if board.count(0)==0 or s!=0:
        # om någon av ovanstående konditioner är True, så är spelet över.
        return True
    #om det är false så är spelet inte över.
    return False

#Loopar 0 till 3, alltså 4 gånger. för att se så att funktionen fungerar. 
for i in range(4):
    # hämta ett random bräde.
    board=getRandomBoard()
    # printa ut brädet.
    showBoard(board)
    # printa resultatet av gameOver funktionen med det nuvarande brädet som input
    print(gameOver(board))
```

```python
# skapa en tom dictionary
newestAdditionTwo = dict()

# funktion som lägger till nya möjliga bräden i newestAdditionTwo dictionaryn
# board är brädet som ska undersökas och who är vilken spelares tur det är
def addPossibleBoards(board, who): 
    # moves hämtar alla möjliga drag för  brädet
    moves = getAllMoves(board)
    # lista med antalet tillåtna drag, där varje element har värdet 1
    V = np.ones(len(moves))
    # normaliserar listan så att alla element summerar till 1
    V = V / V.sum()
    # om spelet inte är över och det finns minst en tom ruta kvar
    if score(board) == 0 and list(board).count(0) != 0:
        # lägg till brädet i newestAdditionTwo med values från V
        newestAdditionTwo[tuple(board)] = V
    # en lista som swapar mellan player1 och player2
    flipMove = [0, 2, 1]
    # för varje tillåtet drag(move) i moves
    for move in moves:
        # gör draget på brädet och få fram det nya brädet
        newBoard = doMove(board, move, who)
        # undersök alla möjliga bräden som kan uppstå efter det nya draget
        addPossibleBoards(newBoard, flipMove[who])

# hämta ett tomt bräde
board = getEmptyBoard()
# undersöker möjliga bräden som spelare 1.
addPossibleBoards(board, 1)
# skriv ut antalet bräden från newestAdditionTwo
print(len(newestAdditionTwo))
```

```python
#koden simulerar ett game mellan två spelare med policies definerade i policyA och B. Under spelets gång så 
#används reinforcement learning q-learning till att uppdatera newestpolicy utifrån game:ets "outcome". 
#Sedan så uppdaterar den och sparas in i filen newestpolicy
def playTwoPolicies(policyA,policyB,verbose=False):

    # Initialiserar dictionary för att spara learning rates för olika strats
    WinStrat_player1 = dict() 
    WinStrat_player2 = dict() 
    TieStrat_player1 = dict() 
    TieStrat_player2 = dict() 
    
    # Definerar en lista för att spara moves för spelarna
    flipMove=[0,2,1]
     #vilken spelare
    who=1
   #tom bräda
    board=[0,0,0,0,0,0,0,0,0]
    done=False
    
    # Definerar learning rates för olika game outcomes
    wonGame = 0.10
    tiedGame = 0.05 # hälften så mycket som för en win eftersom det inte är det värsta outcomen
    lostGame = 0
    
    while not done:
        if verbose:
            showBoard(board)
        
        # hämtar available moves för current player
        moves=getAllMoves(board)
        
        # Väljer policy för current player
        if who==1:
            p=policyA[tuple(board)]
        else:
            p=policyB[tuple(board)]
        
        if verbose:
            print(p)

        # Normalize sannorlikheter
        p/=p.sum()
        
        # väljer move utifrån sannorlikhet distribution
        choice=numpy.random.choice(moves,p=p) #[0,1,0,0,0,0,0,0,0]
        
        # Spara det valda move:et för current player
        # Kontrollerar om det är spelare 1 eller spelare 2
        if who == 1:
            # För spelare 1 : skapar arrayer för att lagra learning rate för wins och ties för varje möjligt drag
            move_player1 = zeros(len(moves))
            winLearningrate1 = zeros(len(moves))
            tieLearningrate1 = zeros(len(moves))

            # Hämta indexet för det valda draget och ställ in motsvarande värde i move_player1-arrayen till 1
            newIndex = moves.index(choice)
            move_player1[newIndex] = 1 

            # Lagrar utfallen för wonGame och tiedGame i motsvarande arrayer för det valda draget
            winLearningrate1[newIndex] = wonGame 
            tieLearningrate1[newIndex] = tiedGame

            # Deep copy de vinnande och tying learning rate för senare användning
            winnerMove1 = copy.deepcopy(winLearningrate1)
            tieMove1 = copy.deepcopy(tieLearningrate1) 

            # Uppdatera vinst- och tying learning rates för spelets nuvarande tillstånd
            WinStrat_player1[tuple(board)] = winnerMove1 
            TieStrat_player1[tuple(board)] = tieMove1
        else:
            # För spelare 2 : skapar arrayer för att lagra learning rate för wins och ties för varje möjligt drag
            winLearningrate2 = zeros(len(moves))
            tieLearningrate2 = zeros(len(moves))
            move_player2 = zeros(len(moves))

            # Hämta indexet för det valda draget och ställ in motsvarande värde i move_player2-arrayen till 1
            newIndex = moves.index(choice)
            move_player2[newIndex] = 1

            # Lagrar utfallen för wonGame och TiedGame i motsvarande arrayer för det valda draget
            winLearningrate2[newIndex] = wonGame
            tieLearningrate2[newIndex] = tiedGame 

            # Deep copy de vinnande och tying learning rate för senare användning
            winnerMove2 = copy.deepcopy(winLearningrate2)
            tieMove2 = copy.deepcopy(tieLearningrate2)

            # Uppdatera vinst- och tying learning rates för spelets nuvarande tillstånd
            WinStrat_player2[tuple(board)] = winnerMove2
            TieStrat_player2[tuple(board)] = tieMove2
        
        # gör det valda movet på brädan
        board=doMove(board,choice,who)

        # kollar score efter ett move
        s=score(board)
        
        if len(moves) == 1 or s != 0: # kollar ifall det finns bara 1 move eller om gamet är klart 
        
            if s == 0: # Game tied
                saveTieStrat_player1 = TieStrat_player1 # Sparar player 1:s tying strategy
                saveTieStrat_player2 = TieStrat_player2 # Sparar player 2:s tying strategy, då  game är en tied
                for state in saveTieStrat_player1:
                    QL_Policy[state] += saveTieStrat_player1[state] # Uppdatera Q-learning policyn med player 1:s tying strategy
                    QL_Policy[state] += saveTieStrat_player2[state] # Uppdatera Q-learning policyn med player 2:s tying strategy
            elif s == 1: # Player 1 vinner
                saveWinStrat = WinStrat_player1 # Spara player 1:s winning strategy
                for state in saveWinStrat:
                    QL_Policy[state] += saveWinStrat[state] # Uppdatera Q-learning policyn med player1:s winning strategy
            elif s == 2: # Player 2 vinner
                saveWinStrat = WinStrat_player2 # Spara player 2's winning strategy
                for state in saveWinStrat:
                    QL_Policy[state] += saveWinStrat[state] # Uppdatera Q-learning policyn med player2:s winning strategy

            done = True 
        
    who = flipMove[who] # vilken spelarens tur
    if verbose:
        showBoard(board)
    if s == 0:
        return 0 # Ifall gamet är en tie, returnera 0
    return # Return inget annars

    # Kör 1000 games och uppdaterar Q-learning policyn
    for i in range(1000):
        playTwoPolicies(perfectPolicy, inClassRandomPolicy, verbose=False)

# Spara uppdaterade Q-learning policyn till filen
pickle.dump(newestPolicy, open("newestPolicy.p", "wb"))

# Load den uppdaterade Q-learning policyn från filen
newestPolicyUpdated = pickle.load(open("newestPolicy.p", "rb"))
```

```python
# Denna funktion simulerar ett game av tic tac toe mellan två policies a och b, varav varje policy gör random drag utifrån
# sin policy och vad som är "possiblemoves" alltså de tillgängliga platserna på brädet.
def PlayTwoPoliciesRandom(policyA,policyB,verbose=False):
    # En lista som byter spelare efter varje drag.
    flipMove=[0,2,1]
    # Variabeln som håller reda på vilken spelare som ska göra nästa drag. Spelare 1 börjar.
    who=1
    # En lista som används som brädet.
    board=[0,0,0,0,0,0,0,0,0]
    # håller koll på om spelet är klart eller ej.
    done=False
# funktionen retunerar vinnaren av spelet. 1 = policyA, 2= policyB och 0 = oavgjort. Om verbose är satt till true så 
#visas spelet, false så visas bara vinnaren.

    # Fortsätt loopa tills done=True.
    while not done:
        if verbose:
            showBoard(board)
        moves=getAllMoves(board)
        if who==1:
            # Hämta sannolikheten för varje möjligt drag givet policyA.
            p=policyA[tuple(board)]
        else:
            # Hämta sannolikheten för varje möjligt drag givet policyB.
            p=policyB[tuple(board)]
        if verbose:
            # Visa sannolikheterna för varje möjligt drag.
            print(p)
        #Normalisera dem (sannolikheterna)
        p/=p.sum()
        #Välj ett drag randomly utifrån dem.
        choice=numpy.random.choice(moves,p=p)
        #genomför draget`!
        board=doMove(board,choice,who)
        #checka om spelet är över
        s=score(board)
        #Om det bara finns ett möjligt drag kvar eller om spelet är över-> sätt done=True.
        if len(moves)==1 or s!=0:
            done=True
            # byt spelare
        who=flipMove[who]
    if verbose:
        # Visa brädet en sista gång.
        showBoard(board)
    # Om det blir en tie så returneras 0. 
    if s==0:
        return 0
    # Returnera vinnaren (1 = policyA, 2 = policyB)
    return flipMove[who]
print(PlayTwoPoliciesRandom(inClassRandomPolicy,inClassRandomPolicy,verbose=True))
```

```python
#sampleGames tar in två policies, policyA och polilcyB, och antalet games som ska spelas. Efter varje game håller den reda 
#på vinnaren och returnerar andelen games som blivit draws, vunnits eller förlorats.
def sampleGames(policyA,policyB,nrOfGames=100):
    result=[0,0,0]
    for n in range(nrOfGames):
        winner=PlayTwoPoliciesRandom(policyA,policyB)
        result[winner]+=1
    result=numpy.array(result)
    return result/result.sum()
print(" Tie% - Player O wr% - Player X wr%")
print("----------------------")
print("RvR",sampleGames(inClassRandomPolicy,inClassRandomPolicy,nrOfGames=100000))
print("PvP",sampleGames(perfectPolicy,perfectPolicy,nrOfGames=100000))
print("PvR",sampleGames(perfectPolicy,inClassRandomPolicy,nrOfGames=100000))
print("RvP",sampleGames(inClassRandomPolicy,perfectPolicy,nrOfGames=100000))
print("our QL policy vs Perfectpolicy")
print("NvR",sampleGames(newestPolicyUpdated,perfectPolicy,nrOfGames=100000))
```
